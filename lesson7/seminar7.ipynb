```python
import os
import random

# w, a, r - запись, добавление, чтение

# my_file = open('numbers.txt', 'a', encoding='UTF-8')
# my_file.write(input('Введите строку для записи: ') + '\n')
# my_file.close()


# ✔Напишите функцию, которая заполняет файл (добавляет в конец) случайными парами чисел.
# ✔Первое число int, второе - float разделены вертикальной чертой. ✔Минимальное число - -1000, максимальное - +1000.
# ✔Количество строк и имя файла передаются как аргументы функции.

def fill_file_with_random_pairs(count, filename):
  with open(filename, "a") as file:
    for _ in range(count):
      n = int(random.uniform(-1000, 1000))
      f = float(random.uniform(-1000, 1000))
      file.write(f"{n}|{f}\n")


# ✔Напишите функцию, которая генерирует псевдоимена.
# ✔Имя должно начинаться с заглавной буквы, состоять из 4-7 букв, среди которых обязательно должны быть гласные.
# ✔Полученные имена сохраните в файл.


def generate_pseudonym(min_length=4, max_length=7):
  vowels = ["a", "e", "i", "o", "u"]
  length = random.randint(min_length, max_length)
  pseudonym = ""
  for _ in range(length):
    letter = random.choice(["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z"])
    pseudonym += letter
  if not any(letter in pseudonym for letter in vowels):
    pseudonym += random.choice(vowels)
  return pseudonym.title()


def write_pseudonyms():
  count = 100
  filename = "pseudonyms.txt"
  pseudonyms = []
  for _ in range(count):
    pseudonyms.append(generate_pseudonym())
  with open(filename, "w") as file:
    for pseudonym in pseudonyms:
      file.write(pseudonym + "\n")

# write_pseudonyms()

# ✔Создайте функцию, которая создаёт файлы с указанным расширением.
# Функция принимает следующие параметры: ✔расширение ✔минимальная длина случайно сгенерированного имени, по умолчанию 6
# ✔максимальная длина случайно сгенерированного имени, по умолчанию 30 ✔минимальное число случайных байт, записанных в файл,
# по умолчанию 256 ✔максимальное число случайных байт, записанных в файл, по умолчанию 4096 ✔количество файлов, по умолчанию 42
# ✔Имя файла и его размер должны быть в рамках переданного диапазона.


def create_files(extension, min_length=6, max_length=30, min_size=256, max_size=4096, count=42):
  names = []
  for _ in range(count):
    names.append("".join(random.choice("abcdefghijklmnopqrstuvwxyz") for _ in range(random.randint(min_length, max_length))))
  for name in names:
    size = random.randint(min_size, max_size)
    path = os.path.join(os.getcwd(), name + extension)
    with open(path, "wb") as file:
      for _ in range(size):
        file.write(random.randint(0, 255).to_bytes(1, byteorder="big"))

# create_files(".txt", 6, 30, 256, 4096, 42)


def delete_txt_files():
  files = os.listdir()
  txt_files = [file for file in files if file.endswith(".txt")]
  print(txt_files)
  for file in txt_files:
    os.remove(file)

# delete_txt_files()


# with open('names.txt', 'r', encoding='UTF-8') as file:
#     print(file.read().split('\n'))

# with open('names.txt', 'r', encoding='UTF-8') as file:
#     print([row.strip() for row in file.readlines()])


# ✔Напишите функцию, которая открывает на чтение созданные в прошлых задачах файлы с числами и именами.
# ✔Перемножьте пары чисел. В новый файл сохраните имя и произведение: ✔если результат умножения отрицательный,
# сохраните имя записанное строчными буквами и произведение по модулю ✔если результат умножения положительный,
# сохраните имя прописными буквами и произведение округлённое до целого.
# ✔В результирующем файле должно быть столько же строк, сколько в более длинном файле.
# ✔При достижении конца более короткого файла, возвращайтесь в его начало.


def multiply_and_save(numbers_filename, names_filename, results_filename):
  files = os.listdir()
  if numbers_filename not in files or names_filename not in files:
    raise FileNotFoundError("Файлы не найдены")

  with open(numbers_filename, "r") as numbers_file, open(names_filename, "r") as names_file:
    numbers = numbers_file.readlines()
    names = names_file.readlines()

  if len(numbers) != len(names):
    raise ValueError("Количество строк в файлах не совпадает")

  with open(results_filename, "w") as results_file:
    for number, name in zip(numbers, names):
      product = int(number) * int(name)
      if product < 0:
        results_file.write(f"{name.lower()} {abs(product)}\n")
      else:
        results_file.write(f"{name.upper()} {product}\n")


# multiply_and_save('numbers.txt', 'names.txt', 'results.txt')
```
