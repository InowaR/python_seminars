```python
def task1():
  my_list = [1,2,3,4,1,2,3]
  print(list(set(my_list)))
  new_list = []
  for e in my_list:
    if e not in new_list:
      new_list.append(e)
  print(new_list)

# task1()

def task2():
  user = input('Введите данные: ')
  if user.isdigit():
    print('int')
  elif user.count('.') == 1 and user.replace('.', '').replace('-', '').isdigit():
    print('float')
  elif any([c.isupper() for c in user]):
    print('Upper')
  else:
    print('Lower')

# task2()

def task3():
  arr = (2, 'hi', [1,2,3], 13.34, True, (1, 'hi', 3.2), 'hi', 2, None)
  new_dict = {}
  for item in arr:
    t = type(item)
    if t in new_dict:
      new_dict[t].append(item)
    else:
      new_dict[t] = [item]
  print(new_dict)

# task3()

def task4():
  my_list = [1,2,2,3,3,4]
  for i in set(my_list):
    if my_list.count(i) == 2:
      my_list.remove(i)
      my_list.remove(i)
  print(my_list)


# task4()

def task5():
  arr = [1,2,3,7,9,11,13,14,16]
  new_list = []
  for i in range(len(arr)):
    if arr[i] % 2 != 0:
      new_list.append(i)
  print(new_list)

# task5()

def task6():
  my_str = input('Введите строку: ').split()
  my_str.sort()
  print(my_str)
  max_len = len(max(my_str, key=len))
  for num, element in enumerate(my_str, 1):
    print(f'{num}. {element:.>{max_len}}')

# task6()

def task7():
  my_string = input('Введите строку: ')
  my_set = set(my_string)
  my_dict = {}
  # count = 0
  # for i in my_set:
  #   for j in range(len(my_string)):
  #     if my_string[j] == i:
  #       count += 1
  #   my_dict[i] = count
  #   count = 0
  # print(my_dict)
  for i in my_string:
    my_dict[i] = my_dict.get(i, 0) + 1
  print(my_dict)

# task7()

def task8():
  friends_stuff = {
      'Олег': ('палатка', 'топор','еда','пиво'),
      'Игорь': ('палатка', 'вилка','вода','пиво'),
      'Гусь': ('палатка', 'топор','вода','пиво'),
      'Стоун': ('палатка', 'топор','вода','лимонад')
  }
  set1 = set()
  for k in friends_stuff:
    if not set1:
      set1 = set(friends_stuff[k])
    else:
      set1 &= set(friends_stuff[k])
  print('Общие вещи друзей: ', *set1)

  my_tuple = friends_stuff.keys()
  my_set = set()
  for friend in my_tuple:
    my_set = set(friends_stuff[friend])
    for e in [i for i in my_tuple if i != friend]:
      my_set = my_set - set(friends_stuff[e])
    if my_set:
      print('Есть только у ', friend, ': ', *my_set)
  for friend in my_tuple:
    my_set = set()
    to_remove = set(friends_stuff[friend])
    for e in [i for i in my_tuple if i != friend]:
      if not my_set:
        my_set = set(friends_stuff[e])
      else:
        my_set = my_set & set(friends_stuff[e])
    my_set -= to_remove
    if my_set:
      print(friend, 'не взял', *my_set)

task8()

from copy import deepcopy
def example():
  a = 1
  b = a
  a += 1
  print(a)
  print(b)
  arr1 = [1,2,3]
  # arr2 = arr1
  # arr1.append(4)
  # print(arr1)
  # print(arr2)
  # arr2 = arr1.copy() # не работает со вложенными списками
  # arr1.append(4)
  # print(arr1)
  # print(arr2)
  arr2 = deepcopy(arr1)
  arr1.append(4)
  print(arr1)
  print(arr2)
  a = {(1,2,3,(1,2,3)): (1,2,3)}
  print(a)
  # b = {(1,2,3,[1,2,3]): (1,2,3)} #TypeError: unhashable type: 'list'
  # print(b)

example()
```
