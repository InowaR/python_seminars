```python
import math

# üìå –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å –æ–∫—Ä—É–∂–Ω–æ—Å—Ç—å. üìå –ö–ª–∞—Å—Å –¥–æ–ª–∂–µ–Ω –ø—Ä–∏–Ω–∏–º–∞—Ç—å —Ä–∞–¥–∏—É—Å –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞.
# üìå –£ –∫–ª–∞—Å—Å–∞ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –¥–≤–∞ –º–µ—Ç–æ–¥–∞, –≤–æ–∑–≤—Ä–∞—â–∞—é—â–∏–µ –¥–ª–∏–Ω—É –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ –∏ –µ—ë –ø–ª–æ—â–∞–¥—å.

class Circle:
  def __init__(self, radius):
    self.radius = radius

  def length(self):
    return 2 * math.pi * self.radius

  def area(self):
    return math.pi * self.radius**2


# circle = Circle(5)
# print(circle.length())
# print(circle.area())


# üìå –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫. üìå –ö–ª–∞—Å—Å –¥–æ–ª–∂–µ–Ω –ø—Ä–∏–Ω–∏–º–∞—Ç—å –¥–ª–∏–Ω—É –∏ —à–∏—Ä–∏–Ω—É –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞.
# üìå –£ –∫–ª–∞—Å—Å–∞ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –¥–≤–∞ –º–µ—Ç–æ–¥–∞, –≤–æ–∑–≤—Ä–∞—â–∞—é—â–∏–µ –ø–µ—Ä–∏–º–µ—Ç—Ä –∏ –ø–ª–æ—â–∞–¥—å.
# üìå –ï—Å–ª–∏ –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ –ø–µ—Ä–µ–¥–∞—ë—Ç—Å—è —Ç–æ–ª—å–∫–æ –æ–¥–Ω–∞ —Å—Ç–æ—Ä–æ–Ω–∞, —Å—á–∏—Ç–∞–µ–º —á—Ç–æ —É –Ω–∞—Å –∫–≤–∞–¥—Ä–∞—Ç.

class Rectangle:
  def __init__(self, length, width=None):
    self.length = length
    self.width = width if width else length

  def perimeter(self):
    return 2 * self.length + 2 * self.width

  def area(self):
    return self.length * self.width


# rectangle = Rectangle(2,3)
# print(rectangle.perimeter())
# print(rectangle.area())

# rectangle = Rectangle(2)
# print(rectangle.perimeter())
# print(rectangle.area())


# üìå –ù–∞–ø–∏—à–∏—Ç–µ –∫–ª–∞—Å—Å –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —á–µ–ª–æ–≤–µ–∫–µ: –§–ò–û, –≤–æ–∑—Ä–∞—Å—Ç –∏ —Ç.–ø. –Ω–∞ –≤–∞—à –≤—ã–±–æ—Ä.
# üìå –£ –∫–ª–∞—Å—Å–∞ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –º–µ—Ç–æ–¥—ã birthday –¥–ª—è —É–≤–µ–ª–∏—á–µ–Ω–∏—è –≤–æ–∑—Ä–∞—Å—Ç–∞ –Ω–∞ –≥–æ–¥, full_name –¥–ª—è –≤—ã–≤–æ–¥–∞ –ø–æ–ª–Ω–æ–≥–æ –§–ò–û –∏ —Ç.–ø. –Ω–∞ –≤–∞—à –≤—ã–±–æ—Ä.
# üìå –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ —Å–≤–æ–π—Å—Ç–≤–æ –≤–æ–∑—Ä–∞—Å—Ç –Ω–µ–¥–æ—Å—Ç—É–ø–Ω–æ –¥–ª—è –ø—Ä—è–º–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è, –Ω–æ –µ—Å—Ç—å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å —Ç–µ–∫—É—â–∏–π –≤–æ–∑—Ä–∞—Å—Ç.

class Person:
  def __init__(self, firstname, lastname, age):
    self._firstname = firstname
    self._lastname = lastname
    self._age = age

  def age(self):
    return self._age

  def full_name(self):
    return self._firstname, self._lastname

  def birthday(self):
    self._age += 1


person1 = Person('Nikolay', 'Petrov', 30)
person1.age()
person1.birthday()
person1.age()


# üìå –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å –°–æ—Ç—Ä—É–¥–Ω–∏–∫. üìå –í–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å –∫–ª–∞—Å—Å–æ–º —á–µ–ª–æ–≤–µ–∫–∞ –∏–∑ –ø—Ä–æ—à–ª–æ–≥–æ –∑–∞–¥–∞–Ω–∏—è.
# üìå –£ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å: ‚óã —à–µ—Å—Ç–∏–∑–Ω–∞—á–Ω—ã–π –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ü–∏–æ–Ω–Ω—ã–π –Ω–æ–º–µ—Ä ‚óã —É—Ä–æ–≤–µ–Ω—å –¥–æ—Å—Ç—É–ø–∞ –≤—ã—á–∏—Å–ª—è–µ–º—ã–π –∫–∞–∫ –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è —Å—É–º–º—ã —Ü–∏—Ñ—Ä id –Ω–∞ —Å–µ–º—å


class Employee(Person):
  def __init__(self, firstname, lastname, age, id_number):
    super().__init__(firstname, lastname, age)
    if len(str(id_number)) == 6:
      self.id_number = str(id_number)
    elif len(str(id_number)) <= 6:
      self.id_number = str(id_number).zfill(6)
    else:
      self.id_number = str(id_number)[:6]
    self.access_number = sum(int(i) for i in self.id_number) % 7

  def lvl(self):
    return self.id_number


# e = Employee('Nikolay', 'Petrov', 30, 1234)
# e.lvl()
# e.access_number


# üìå –°–æ–∑–¥–∞–π—Ç–µ —Ç—Ä–∏ (–∏–ª–∏ –±–æ–ª–µ–µ) –æ—Ç–¥–µ–ª—å–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤ –∂–∏–≤–æ—Ç–Ω—ã—Ö. –ù–∞–ø—Ä–∏–º–µ—Ä —Ä—ã–±—ã, –ø—Ç–∏—Ü—ã –∏ —Ç.–ø.
# üìå –£ –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –∫–∞–∫ –æ–±—â–∏–µ —Å–≤–æ–π—Å—Ç–≤–∞, –Ω–∞–ø—Ä–∏–º–µ—Ä –∏–º—è, —Ç–∞–∫ –∏ —Å–ø–µ—Ü–∏—Ñ–∏—á–Ω—ã–µ –¥–ª—è –∫–ª–∞—Å—Å–∞.
# üìå –î–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞ —Å–æ–∑–¥–∞–π—Ç–µ –º–µ—Ç–æ–¥, –≤—ã–≤–æ–¥—è—â–∏–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Å–ø–µ—Ü–∏—Ñ–∏—á–Ω—É—é –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞.

class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def __str__(self):
    return f'{self.name}, {self.age}'


class Fish(Animal):
  def __init__(self, name, age, fish_type):
    super().__init__(name, age)
    self.fish_type = fish_type

  # def __str__(self):
  #   return f'{self.name}, {self.age}, {self.fish_type}'

  @staticmethod
  def hello():
    return 'hello'

  @classmethod
  def hi(cls):
    return 'classmethod'


class Bird(Animal):
  def __init__(self, name, age, bird_type):
    super().__init__(name, age)
    self.bird_type = bird_type

  def __str__(self):
    return f'{self.name}, {self.age}, {self.bird_type}'


class Horse(Animal):
  def __init__(self, name, age, horse_type):
    super().__init__(name, age)
    self.horse_type = horse_type

  def __str__(self):
    return f'{self.name}, {self.age}, {self.horse_type}'


fish = Fish('Nick', '12', 'sea')
bird = Bird('John', '1', 'red')
horse = Horse('Mike', '2', 'big')

print(fish)
print(fish.hello())
print(Fish.hello())
print(fish.hi())
print(Fish.hi())
print(bird)
print(horse)
```
