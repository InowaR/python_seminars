```python
# üìå –°–æ–∑–¥–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é-–∑–∞–º—ã–∫–∞–Ω–∏–µ, –∫–æ—Ç–æ—Ä–∞—è –∑–∞–ø—Ä–∞—à–∏–≤–∞–µ—Ç –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞: ‚óã –æ—Ç 1 –¥–æ 100 –¥–ª—è –∑–∞–≥–∞–¥—ã–≤–∞–Ω–∏—è, ‚óã –æ—Ç 1 –¥–æ 10 –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–ø—ã—Ç–æ–∫ üìå
# –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è —á–µ—Ä–µ–∑ –∫–æ–Ω—Å–æ–ª—å –ø—Ä–æ—Å–∏—Ç —É–≥–∞–¥–∞—Ç—å –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –∑–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –ø–æ–ø—ã—Ç–æ–∫.

from random import randint as ri

MIN_LIMIT = 1
MAX_LIMIT = 100
MIN_COUNT = 1
MAX_COUNT = 10


def guess_number(number: int, count: int):
    user_number = number if MIN_LIMIT <= number <= MAX_LIMIT else ri(MIN_LIMIT, MAX_LIMIT)
    user_count = count if MIN_COUNT <= count <= MAX_COUNT else ri(MIN_COUNT, MAX_COUNT)

    def inner():
        nonlocal user_count
        while user_count:
            guess_num = int(input(f'–£–≥–∞–¥–∞–π—Ç–µ —á–∏—Å–ª–æ –æ—Ç {MIN_LIMIT} –¥–æ {MAX_LIMIT}: '))
            if guess_num == user_number:
                print(f'–£—Ä–∞, —Ç—ã –ø–æ–±–µ–¥–∏–ª! –≠—Ç–æ —á–∏—Å–ª–æ {user_number}')
                return
            elif guess_num < user_number:
                print('–ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –±–æ–ª—å—à–µ!')
            else:
                print('–ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –º–µ–Ω—å—à–µ!')
            user_count -= 1
        print(f'–£–≤—ã! –¢—ã –ø—Ä–æ–∏–≥—Ä–∞–ª! –≠—Ç–æ –±—ã–ª–æ —á–∏—Å–ª–æ {user_number}!')

    return inner


# if __name__ == '__main__':
#     game = guess_number(169, 7)
#     game()



# –° –¥–µ–∫–æ—Ä–∞—Ç–æ—Ä–æ–º

MIN_LIMIT = 1
MAX_LIMIT = 100
MIN_COUNT = 1
MAX_COUNT = 10


def game_rules(func):
    def inner(number, count):
        user_number = number if MIN_LIMIT <= number <= MAX_LIMIT else ri(MIN_LIMIT, MAX_LIMIT)
        user_count = count if MIN_COUNT <= count <= MAX_COUNT else ri(MIN_COUNT, MAX_COUNT)
        func(user_number, user_count)
    return inner


@game_rules
def guess_game(user_number, user_count):
    while user_count:
        guess_num = int(input(f'–£–≥–∞–¥–∞–π—Ç–µ —á–∏—Å–ª–æ –æ—Ç {MIN_LIMIT} –¥–æ {MAX_LIMIT}: '))
        if guess_num == user_number:
            print(f'–£—Ä–∞, —Ç—ã –ø–æ–±–µ–¥–∏–ª! –≠—Ç–æ —á–∏—Å–ª–æ {user_number}')
            return
        elif guess_num < user_number:
            print('–ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –±–æ–ª—å—à–µ!')
        else:
            print('–ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –º–µ–Ω—å—à–µ!')
        user_count -= 1
    print(f'–£–≤—ã! –¢—ã –ø—Ä–æ–∏–≥—Ä–∞–ª! –≠—Ç–æ –±—ã–ª–æ —á–∏—Å–ª–æ {user_number}!')



# if __name__ == '__main__':
#     guess_game(17, 7)



# üìå –ù–∞–ø–∏—à–∏—Ç–µ –¥–µ–∫–æ—Ä–∞—Ç–æ—Ä, –∫–æ—Ç–æ—Ä—ã–π —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç –≤ json —Ñ–∞–π–ª –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–µ–∫–æ—Ä–∏—Ä—É–µ–º–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –∫–æ—Ç–æ—Ä—ã–π –æ–Ω–∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç.
# –ü—Ä–∏ –ø–æ–≤—Ç–æ—Ä–Ω–æ–º –≤—ã–∑–æ–≤–µ —Ñ–∞–π–ª –¥–æ–ª–∂–µ–Ω —Ä–∞—Å—à–∏—Ä—è—Ç—å—Å—è, –∞ –Ω–µ –ø–µ—Ä–µ–∑–∞–ø–∏—Å—ã–≤–∞—Ç—å—Å—è.
# üìå –ö–∞–∂–¥—ã–π –∫–ª—é—á–µ–≤–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ –∫–∞–∫ –æ—Ç–¥–µ–ª—å–Ω—ã–π –∫–ª—é—á json —Å–ª–æ–≤–∞—Ä—è.
# üìå –î–ª—è –¥–µ–∫–æ—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –Ω–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –º–æ–∂–µ—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∫–∞–∫ –ø–æ–∑–∏—Ü–∏–æ–Ω–Ω—ã–µ, —Ç–∞–∫ –∏ –∫–ª—é—á–µ–≤—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã.
# üìå –ò–º—è —Ñ–∞–π–ª–∞ –¥–æ–ª–∂–Ω–æ —Å–æ–≤–ø–∞–¥–∞—Ç—å —Å –∏–º–µ–Ω–µ–º –¥–µ–∫–æ—Ä–∏—Ä—É–µ–º–æ–π —Ñ—É–Ω–∫—Ü–∏–∏.


import json

def log_results(function):
    def wrapper(*args, **kwargs):
        result = function(*args, **kwargs)
        function_name = function.__name__
        parameters = {
            "args": args,
            "kwargs": kwargs,
            "result": result
        }
        with open(function_name + ".json", "a") as f:
            f.write(json.dumps(parameters) + "\n")
        return function(*args, **kwargs)
    return wrapper

@log_results
def add(a, b):
    return a + b

print(add(1, 2))



import json
import os.path
from functools import wraps


def json_decor(func):
    @wraps(func)
    def inner(*args, **kwargs):
        file_name = f'{func.__name__}.json'
        if os.path.exists(file_name):
            with open(file_name, 'r', encoding='UTF-8') as file:
                result = json.load(file)
        else:
            result = {}
        primary_key = func(*args, **kwargs)
        primary_value = {}
        primary_value['args'] = args
        primary_value['kwargs'] = {}

        for key, value in kwargs.items():
            primary_value['kwargs'][key] = value
        result[primary_key] = primary_value
        with open(file_name, 'w', encoding='UTF-8') as file:
            json.dump(result, file, indent=4, ensure_ascii=False)

    return inner


@json_decor
def some_func(a, b):
    return a + b


# –î–µ–∫–æ—Ä–∞—Ç–æ—Ä —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏

def my_decorator(function, name="Decorator", prefix="", suffix=""):
    def wrapper(*args, **kwargs):
        print(f"{prefix} {name} {function.__name__} {suffix}")
        return function(*args, **kwargs)
    return wrapper

@my_decorator(name="My decorator", prefix="–í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏: ", suffix="")
def add(a, b):
    return a + b


# from random import randint as ri
# from main_3 import json_decor
# from main_4 import deco_with_params
# from functools import wraps

MIN_LIMIT = 1
MAX_LIMIT = 100
MIN_COUNT = 1
MAX_COUNT = 10


def game_rules(func):
    @wraps(func)
    def inner(number, count):
        user_number = number if MIN_LIMIT <= number <= MAX_LIMIT else ri(MIN_LIMIT, MAX_LIMIT)
        user_count = count if MIN_COUNT <= count <= MAX_COUNT else ri(MIN_COUNT, MAX_COUNT)
        return func(user_number, user_count)

    return inner


# @deco_with_params(1)
@json_decor
@game_rules
def guess_game(user_number, user_count):
    while user_count:
        guess_num = int(input(f'–£–≥–∞–¥–∞–π—Ç–µ —á–∏—Å–ª–æ –æ—Ç {MIN_LIMIT} –¥–æ {MAX_LIMIT}: '))
        if guess_num == user_number:
            print(f'–£—Ä–∞, —Ç—ã –ø–æ–±–µ–¥–∏–ª! –≠—Ç–æ —á–∏—Å–ª–æ {user_number}')
            return f'–ü–æ–±–µ–¥–∞! {user_number} {user_count}'
        elif guess_num < user_number:
            print('–ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –±–æ–ª—å—à–µ!')
        else:
            print('–ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –º–µ–Ω—å—à–µ!')
        user_count -= 1
    print(f'–£–≤—ã! –¢—ã –ø—Ä–æ–∏–≥—Ä–∞–ª! –≠—Ç–æ –±—ã–ª–æ —á–∏—Å–ª–æ {user_number}!')
    return f'–ü—Ä–æ–∏–≥—Ä—ã—à! {user_number} {user_count}'


if __name__ == '__main__':
    guess_game(102, 7)
```
