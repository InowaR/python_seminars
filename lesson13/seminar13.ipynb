```python
# üìå –°–æ–∑–¥–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –∑–∞–ø—Ä–∞—à–∏–≤–∞–µ—Ç —á–∏—Å–ª–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ –æ–Ω –Ω–µ –≤–≤–µ–¥—ë—Ç —Ü–µ–ª–æ–µ –∏–ª–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ.
# üìå –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–π—Ç–µ –Ω–µ —á–∏—Å–ª–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –∫–∞–∫ –∏—Å–∫–ª—é—á–µ–Ω–∏—è.

def request_number():
  while True:
    try:
      number = input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ: ")
      number = int(number)
      return number
    except ValueError:
      try:
        number = float(number)
        return number
      except ValueError:
        print("–í–≤–µ–¥–µ–Ω–æ –Ω–µ–≤–µ—Ä–Ω–æ–µ —á–∏—Å–ª–æ. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–≤–µ–¥–∏—Ç–µ —Ü–µ–ª–æ–µ –∏–ª–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ.")


# number = request_number()
# print(f"–í–≤–µ–¥–µ–Ω–æ —á–∏—Å–ª–æ: {number}")


# üìå –°–æ–∑–¥–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é –∞–Ω–∞–ª–æ–≥ get –¥–ª—è —Å–ª–æ–≤–∞—Ä—è. üìå –ü–æ–º–∏–º–æ —Å–∞–º–æ–≥–æ —Å–ª–æ–≤–∞—Ä—è —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∫–ª—é—á –∏ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é.
# üìå –ü—Ä–∏ –æ–±—Ä–∞—â–µ–Ω–∏–∏ –∫ –Ω–µ—Å—É—â–µ—Å—Ç–≤—É—é—â–µ–º—É –∫–ª—é—á—É —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å –¥–µ—Ñ–æ–ª—Ç–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ. üìå –†–µ–∞–ª–∏–∑—É–π—Ç–µ —Ä–∞–±–æ—Ç—É —á–µ—Ä–µ–∑ –æ–±—Ä–∞–±–æ—Ç–∫—É –∏—Å–∫–ª—é—á–µ–Ω–∏–π.


def get_from_dict(dict_, key, default=None):
  try:
    return dict_[key]
  except KeyError:
    return default


# my_dict = {"key1": "value1", "key2": "value2"}

# value1 = get_from_dict(my_dict, "key1")
# print(f"value1: {value1}")

# value2 = get_from_dict(my_dict, "key3")
# print(f"value2: {value2}")


# üìå –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å —Å –±–∞–∑–æ–≤—ã–º –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º –∏ –¥–æ—á–µ—Ä–Ω–∏–µ –∫–ª–∞—Å—Å—ã–∏—Å–∫–ª—é—á–µ–Ω–∏—è: ‚óã –æ—à–∏–±–∫–∞ —É—Ä–æ–≤–Ω—è, ‚óã –æ—à–∏–±–∫–∞ –¥–æ—Å—Ç—É–ø–∞.


class MyError(Exception):
  def __init__(self, msg, message):
    self.message = message
    self.msg = msg

  def __str__(self):
    return f'{self.msg}: {self.message}'


class LevelError(MyError):
  def __init__(self, message):
    super().__init__('–û—à–∏–±–∫–∞ —É—Ä–æ–≤–Ω—è', message)


class AccessError(MyError):
  def __init__(self, message):
    super().__init__('–û—à–∏–±–∫–∞ –¥–æ—Å—Ç—É–ø–∞', message)

# raise AccessError('–Ω–µ —Ö–≤–∞—Ç–∞–µ—Ç –ø—Ä–∞–≤')

# –í—Å–ø–æ–º–∏–Ω–∞–µ–º –∑–∞–¥–∞—á—É –∏–∑ —Å–µ–º–∏–Ω–∞—Ä–∞ 8 –ø—Ä–æ —Å–µ—Ä–∏–∞–ª–∏–∑–∞—Ü–∏—é –¥–∞–Ω–Ω—ã—Ö, –≥–¥–µ –≤ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ–º —Ü–∏–∫–ª–µ –∑–∞–ø—Ä–∞—à–∏–≤–∞–ª–∏ –∏–º—è,
# –ª–∏—á–Ω—ã–π –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä –∏ —É—Ä–æ–≤–µ–Ω—å –¥–æ—Å—Ç—É–ø–∞ (–æ—Ç 1 –¥–æ 7) —Å–æ—Ö—Ä–∞–Ω—è—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –≤ JSON —Ñ–∞–π–ª.
# –ù–∞–ø–∏—à–∏—Ç–µ –∫–ª–∞—Å—Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è, –∫–æ—Ç–æ—Ä—ã–π —Ö—Ä–∞–Ω–∏—Ç —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ –≤ —Å–≤–æ–π—Å—Ç–≤–∞—Ö —ç–∫–∑–µ–º–ø–ª—è—Ä–∞.
# –û—Ç–¥–µ–ª—å–Ω–æ –Ω–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏–∑ JSON —Ñ–∞–π–ª–∞ –∏ —Ñ–æ—Ä–º–∏—Ä—É–µ—Ç –º–Ω–æ–∂–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.



import json

def create_users_json(file_path, users):
    data = []
    for user in users:
        user_data = {
            "name": user.name,
            "id": user.id,
            "access_level": user.access_level,
        }
        data.append(user_data)

    with open(file_path, "w", encoding='UTF-8') as f:
        json.dump(data, f, indent=4)

    print(f"–§–∞–π–ª {file_path} —É—Å–ø–µ—à–Ω–æ —Å–æ–∑–¥–∞–Ω!")



class User:
    def __init__(self, name, id, access_level):
        self.name = name
        self.id = id
        self.access_level = access_level

    def __repr__(self):
        return f"User(name={self.name}, id={self.id}, access_level={self.access_level})"



def read_users_from_json(file_path):
    with open(file_path, "r", encoding='UTF-8') as f:
        data = json.load(f)

    users = []
    for user_data in data:
        user = User(
            name=user_data["name"],
            id=user_data["id"],
            access_level=user_data["access_level"],
        )
        users.append(user)

    return users



# users = [
#     User(name="–ò–≤–∞–Ω", id=123, access_level=3),
#     User(name="–ú–∞—Ä–∏—è", id=456, access_level=5),
#     User(name="–ü–µ—Ç—Ä", id=789, access_level=7),
# ]

# file_path = "users.json"

# create_users_json(file_path, users)



# file_path = "users.json"

# users = read_users_from_json(file_path)

# for user in users:
#     print(user)


# üìå –î–æ—Ä–∞–±–æ—Ç–∞–µ–º –∑–∞–¥–∞—á–∏ 3 –∏ 4. –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å –ø—Ä–æ–µ–∫—Ç–∞, –∫–æ—Ç–æ—Ä—ã–π –∏–º–µ–µ—Ç —Å–ª–µ–¥—É—é—â–∏–µ –º–µ—Ç–æ–¥—ã:
# üìå –∑–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö (—Ñ—É–Ω–∫—Ü–∏—è –∏–∑ –∑–∞–¥–∞–Ω–∏—è 4) üìå –≤—Ö–æ–¥ –≤ —Å–∏—Å—Ç–µ–º—É - —Ç—Ä–µ–±—É–µ—Ç —É–∫–∞–∑–∞—Ç—å –∏–º—è –∏ id –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.
# –î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞–ª–∏—á–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –º–Ω–æ–∂–µ—Å—Ç–≤–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –º–∞–≥–∏—á–µ—Å–∫–∏–π –º–µ—Ç–æ–¥ –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.
# –ï—Å–ª–∏ —Ç–∞–∫–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –Ω–µ—Ç, –≤—ã–∑—ã–≤–∞–π—Ç–µ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ –¥–æ—Å—Ç—É–ø–∞.
# –ê –µ—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –µ—Å—Ç—å, –ø–æ–ª—É—á–∏—Ç–µ –µ–≥–æ —É—Ä–æ–≤–µ–Ω—å –∏–∑ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.
# üìå –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è. –ï—Å–ª–∏ —É—Ä–æ–≤–µ–Ω—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –º–µ–Ω—å—à–µ, —á–µ–º –≤–∞—à —É—Ä–æ–≤–µ–Ω—å,
# –≤—ã–∑—ã–≤–∞–π—Ç–µ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ —É—Ä–æ–≤–Ω—è –¥–æ—Å—Ç—É–ø–∞.


import json

class AccessDeniedError(Exception):
    def __init__(self, message="–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω"):
        super().__init__(message)

    def __str__(self):
        return self.message


class User:
    def __init__(self, name, id, access_level):
        self.name = name
        self.id = id
        self.access_level = access_level

    def login(self, project):
        if self.access_level not in project.allowed_access_levels:
            raise AccessDeniedError(f"–£—Ä–æ–≤–µ–Ω—å –¥–æ—Å—Ç—É–ø–∞ '{self.access_level}' –Ω–µ —Ä–∞–∑—Ä–µ—à–µ–Ω –¥–ª—è –ø—Ä–æ–µ–∫—Ç–∞ '{project.name}'")
        return self.access_level


class Project:
    def __init__(self, name, allowed_access_levels):
        self.name = name
        self.allowed_access_levels = allowed_access_levels
        self.users = []

    def add_user(self, user):
        self.users.append(user)

    def login(self, user_name, user_id):
        user = next((u for u in self.users if u.name == user_name and u.id == user_id), None)
        if user is None:
            raise AccessDeniedError("–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –Ω–µ –Ω–∞–π–¥–µ–Ω")
        return user.login(self)


# Example usage
project = Project("–ú–æ–π –ø—Ä–æ–µ–∫—Ç", ["admin"])
user1 = User("–ò–≤–∞–Ω", 123, "admin")
user2 = User("–ü–µ—Ç—Ä", 456, "guest")

# Add users to the project
project.add_user(user1)
project.add_user(user2)

print(project.login(user1.name, user1.id))
print(project.login(user2.name, user2.id))
```
